```python
import pandas as pd
import ast


def convert_to_dict(row):
    try:
        return ast.literal_eval(row) if pd.notnull(row) else {}
    except (ValueError, SyntaxError):
        return {}


attributes_df['BusinessParking'] = attributes_df['BusinessParking'].apply(convert_to_dict)


df_parking = attributes_df['BusinessParking'].apply(pd.Series)
```

```python
df_parking
```

```python
def custom_transform(row):
    if row.isnull().all():
        return 0
    elif row.any():
        return 1
    else:
        return -1


transformed_series = df_parking.apply(custom_transform, axis=1)

df_parking['transformed'] = transformed_series
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have DataFrames named 'result2019', 'result2020', and 'result2021' with columns: business_id, date, count, mean

# Create a list of DataFrames
dfs = [result2019, result2020, result2021]

# Create subplots for 'count' and 'mean'
fig, axs = plt.subplots(2, 3, figsize=(18, 8))

for i, df in enumerate(dfs):
    # Calculate the 'count' and 'mean' distribution
    count_mean_distribution = df.groupby(['business_id', 'year'])[['count', 'mean']].mean().reset_index()

    # Calculate the median values
    median_count = count_mean_distribution['count'].median()
    median_mean = count_mean_distribution['mean'].median()

    # Plot 'count' distribution with median line
    axs[0, i].hist(count_mean_distribution['count'], bins=20, color='skyblue', edgecolor='black')
    axs[0, i].axvline(x=median_count, color='red', linestyle='dashed', linewidth=2, label=f'Median: {median_count:.2f}')
    axs[0, i].set_xlabel('Count')
    axs[0, i].set_ylabel('Frequency')
    axs[0, i].set_title(f'Review Count Distribution ({df["year"].iloc[0]})')
    axs[0, i].legend()

    # Plot 'mean' distribution with median line
    axs[1, i].hist(count_mean_distribution['mean'], bins=20, color='lightcoral', edgecolor='black')
    axs[1, i].axvline(x=median_mean, color='blue', linestyle='dashed', linewidth=2, label=f'Median: {median_mean:.2f}')
    axs[1, i].set_xlabel('Mean Stars')
    axs[1, i].set_ylabel('Frequency')
    axs[1, i].set_title(f'Mean Stars Distribution ({df["year"].iloc[0]})')
    axs[1, i].legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

colors = [
    'blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'purple',
    'orange', 'brown','lime', 'olive', 'gray', 'indigo', 'violet',
]

columns_to_normalize = ['Population Staying at Home', 'Population Not Staying at Home',
                        'Number of Trips <1', 'Number of Trips 1-3',
                        'Number of Trips 3-5', 'Number of Trips 5-10']


for column in columns_to_normalize:
    max_value = Santa_distance[column].max()
    min_value = Santa_distance[column].min()
    Santa_distance[column] = (Santa_distance[column] - min_value) / (max_value - min_value)


plt.figure(figsize=(14, 7))
for i, column in enumerate(columns_to_normalize):
    rolled_series = Santa_distance[column].rolling(window=30, min_periods=1).mean()
    plt.plot(Santa_distance.index, rolled_series, label=column, color=colors[i % len(colors)])

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.tight_layout()
plt.title('Normalized Trends over Time')
plt.xlabel('Date')
plt.ylabel('Normalized Value')
plt.show()
```
