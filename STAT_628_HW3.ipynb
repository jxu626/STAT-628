```python
import pandas as pd
```

```python
import json

# 假设你的JSON数据保存在名为business_info.json的文件中
filename = 'yelp_Fall2023/business.json'

# 打开文件并加载JSON数据
business = []

# 打开文件并逐行读取
with open(filename, 'r') as file:
    for line in file:
        # 将每行解析为 JSON 对象并添加到列表中
        json_obj = json.loads(line)
        business.append(json_obj)

# 将列表转换为 DataFrame
df = pd.DataFrame(business)
```

```python
df.head()
```

```python
# 假设 'df' 是你的 DataFrame，并且 'categories' 列包含了种类字符串
# 先将 'categories' 字段拆分为列表
categories_series = df['categories'].str.split(', ')

# 展平这个列表，以便每个种类都有自己的行
flattened_categories = categories_series.explode()

# 找出唯一的种类
unique_categories = flattened_categories.unique()

# 计算唯一种类的数量
num_unique_categories = len(unique_categories)

print(f"总共有 {num_unique_categories} 种不同的种类。")
```

```python
# 筛选出 'categories' 列包含 "Restaurants" 的行
Shopping_df = df[df['categories'].str.contains('Shopping', na=False)]

# 显示结果
Shopping_df
```

```python
# 筛选出 'categories' 列包含 "Restaurants" 的行
restaurants_df = df[df['categories'].str.contains('Restaurants', na=False)]

# 显示结果
restaurants_df
```

```python
# 在已筛选的 'restaurants_df' DataFrame 中进一步筛选出 'state' 列值为 "LA" 的行
la_restaurants_df = restaurants_df[restaurants_df['state'] == 'LA']

# 显示结果
la_restaurants_df
```

```python
print(la_restaurants_df['attributes'][0:1])
```

```python
# 假设你的JSON数据保存在名为business_info.json的文件中
filename = 'yelp_Fall2023/review.json'

# 打开文件并加载JSON数据
review = []

# 打开文件并逐行读取
with open(filename, 'r') as file:
    for line in file:
        # 将每行解析为 JSON 对象并添加到列表中
        json_obj = json.loads(line)
        review.append(json_obj)

# 将列表转换为 DataFrame
dt = pd.DataFrame(review)
```

```python
dt.head()
```

```python
# 筛选出 'dt' DataFrame 中 'business_id' 存在于 'la_restaurants_df' 中的行
filtered_dt_df = dt[dt['business_id'].isin(la_restaurants_df['business_id'])]

# 显示结果
filtered_dt_df
```

```python
filtered_dt_df['date'] = pd.to_datetime(filtered_dt_df['date'])

# 提取年、月、日到新的列中
filtered_dt_df['year'] = filtered_dt_df['date'].dt.year
filtered_dt_df['month'] = filtered_dt_df['date'].dt.month
filtered_dt_df['day'] = filtered_dt_df['date'].dt.day

filtered_dt_df[['date', 'year', 'month', 'day']]
```
